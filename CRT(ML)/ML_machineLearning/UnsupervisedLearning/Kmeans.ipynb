```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib as jb
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# # unlabelled data for k means clustering
# dt= np.random.rand(100,2)
# print(dt)
```

```python
x,y = make_blobs(n_samples=500, n_features=2, centers=4, cluster_std=1.5, random_state=42)
```

```python
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans.fit(x)
```

```python
print(kmeans.score(x))
```

```python
centers = kmeans.cluster_centers_
print(centers)
```

```python
labels = kmeans.labels_
labels
```

```python
# plotting graph of the clusters
plt.scatter(x[:,0], x[:,1], c=labels, cmap='viridis', marker='o',edgecolor='k') # scatter plot of the clusters
plt.scatter(centers[:,0], centers[:,1], s=100, c='red', marker='*', edgecolor='k', label='Centers') # plot the centers
plt.title('K Means Clustering') # title of the graph
plt.xlabel('Feature 1') # x label
plt.ylabel('Feature 2') # y label
plt.legend()
plt.show()
```
