# Maping

```python
brand_map = {
    "ASUS": 1,
    "Lenovo": 2,
    "acer": 3,
    "Avita": 4,
    "HP": 5,
    "DELL": 6,
    "MSI": 7,
    "APPLE": 8,
}
processor_brand_map = {"Intel": 1, "AMD": 2, "M1": 3}
processor_name_map = {
    "M1": 1,
    "Celeron Dual": 2,
    "Core i3": 3,
    "Core i5": 5,
    "Core i7": 7,
    "Core i9": 9,
    "Pentium Quad": 4,
    "Ryzen 3": 33,
    "Ryzen 5": 55,
    "Ryzen 7": 77,
    "Ryzen 9": 99,
}
processor_gnrtn_map = {
    "Not Available": 0,
    "4th": 4,
    "7th": 7,
    "8th": 8,
    "9th": 9,
    "10th": 10,
    "11th": 11,
    "12th": 12,
}
ram_gb_map = {"4 GB": 4, "8 GB": 8, "16 GB": 16, "32 GB": 32}
ram_type_map = {"DDR4": 44, "LPDDR4": 244, "LPDDR4X": 2447, "DDR5": 45, "DDR3": 43, "LPDDR3": 243}
ssd_map = {
    "0 GB": 0,
    "128 GB": 128,
    "256 GB": 256,
    "512 GB": 512,
    "1024 GB": 1024,
    "2048 GB": 2048,
    "3072 GB": 3072,
}
hdd_map = {"0 GB": 0, "512 GB": 512, "1024 GB": 1024, "2048 GB": 2048}
os_map = {"Windows": 1, "DOS": 2, "Mac": 3}
os_bit_map = {"64-bit": 64, "32-bit": 32}
graphic_card_gb_map = {"0 GB": 0, "2 GB": 2, "4 GB": 4, "6 GB": 6, "8 GB": 8}
weight_map = {"Casual": 1, "ThinNlight": 2, "Gaming": 3}
rating_map = {"1 star": 1, "2 stars": 2, "3 stars": 3, "4 stars": 4, "5 stars": 5}
```

mapping

```python
lp["brand"] = lp["brand"].map(brand_map)
lp["processor_brand"] = lp["processor_brand"].map(processor_brand_map)
lp["processor_name"] = lp["processor_name"].map(processor_name_map)
lp["processor_gnrtn"] = lp["processor_gnrtn"].map(processor_gnrtn_map)
lp["ram_gb"] = lp["ram_gb"].map(ram_gb_map)
lp["ram_type"] = lp["ram_type"].map(ram_type_map)
lp["ssd"] = lp["ssd"].map(ssd_map)
lp["hdd"] = lp["hdd"].map(hdd_map)
lp["os"] = lp["os"].map(os_map)
lp["os_bit"] = lp["os_bit"].map(os_bit_map)
lp["graphic_card_gb"] = lp["graphic_card_gb"].map(graphic_card_gb_map)
lp["weight"] = lp["weight"].map(weight_map)
lp["rating"] = lp["rating"].map(rating_map)
```
