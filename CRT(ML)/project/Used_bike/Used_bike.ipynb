```python
pred
```

```python
pred.shape
```

```python
print("mean_squared_error: ",mean_squared_error(y_test, pred))
```

MAE

```python
print("mean_absolute_error: ",mean_absolute_error(y_test, pred))
```

RMSE

```python
print("r2_score: ",r2_score(y_test, pred))
```

```python
# by using math lib
print("by using math: ", )
```

```python
print("by using numpy: ", np.sqrt(mean_squared_error(y_test, pred)))
```

Hubber Loss

```python
# Calculate Huber loss manually
delta = 1.0  # You can adjust the delta value as needed
huber_loss = np.mean(
    np.where(
        np.abs(y_test - pred) <= delta,
        0.5 * (y_test - pred) ** 2,
        delta * (np.abs(y_test - pred) - 0.5 * delta),
    )
)
print("Huber Loss:", huber_loss)
```
