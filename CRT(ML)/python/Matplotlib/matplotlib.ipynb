```python
x=np.array([0,3,5,7,8,10])
y=np.array([1,4,6,8,9,10])
c=np.array(["red","green","blue","yellow","purple","gray"])
plt.scatter(x,y,c=c)
plt.show()
```

```python
x=np.array([0,3,5,7,8,10])
y=np.array([1,4,6,8,9,10])
size=np.array([100,40,20,150,10,30])
c=np.array(["red","green","blue","yellow","purple","gray"])
plt.scatter(x,y,s=size,c=c,alpha=0.5)
plt.show()
```

```python
x=np.array([0,3,5,7,8,10])
y=np.array([1,4,6,8,9,10])
size=np.array([100,40,20,150,10,30])
c=np.array(["red","green","blue","yellow","purple","gray"])
plt.scatter(x,y,s=size,c=c,alpha=0.5,cmap='nipy_spectral')
plt.colorbar()
plt.show()
```
