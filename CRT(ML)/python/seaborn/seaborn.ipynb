```python
#  lineplot in sns total bill
sns.lineplot(x="total_bill", y="tip", data=data) # lineplot of total bill and tip
```

```python
sns.scatterplot(x="total_bill", y="tip", data=data) # scatterplot of total bill and tip
```

```python
sns.histplot(x="total_bill", data=data) # histogram of total bill
```

```python
sns.kdeplot(data['total_bill'])
```

catplot

```python
sns.catplot(x="total_bill", y="tip", data=data) # catplot of day and total bill
```

```python
data['time'].unique()
```

```python
sns.relplot(x="total_bill", y="tip", hue="time", data=data) # scatterplot with hue of time
```

```python
sns.relplot(x="total_bill", y="tip", hue="time",style="smoker", data=data)  
# scatterplot with hue of time
```

```python
data["smoker"].value_counts() # Count the number of smokers and non-smokers
```

```python
# seprate lunch grapf and dinner graph
sns.relplot(x="total_bill", y="tip", hue="time", style="smoker", data=data,kind='scatter',col="smoker")
# scatterplot with hue of time
```

```python
sns.relplot(x='total_bill', y='tip', hue='time', style='smoker', data=data, kind='scatter', row='time') # scatter plot with row of time
```

```python
data['day'].unique() # unique values of day
```

```python
sns.relplot(
    x="total_bill",
    y="tip",
    hue="day",
    data=data,
    kind="scatter",
    col="day",
)  #
```

```python
sns.relplot(
    x="total_bill",
    y="tip",
    hue="day",
    data=data,
    kind="line",
    col="day",
)  #
```
